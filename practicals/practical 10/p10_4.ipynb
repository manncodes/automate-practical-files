# 1) Laptop price scraper for flipkart

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

import time
```

```python
laptops_data =[]
BASE_URL = 'https://www.flipkart.com/laptops/pr?sid=6bo%2Cb5g&page='
```

```python
def scrape_page(page: int,BASE_URL: str) -> None:
    req = requests.get(BASE_URL + str(page))
    soup = BeautifulSoup(req.text, "html.parser")

    laptop_divs = soup.find_all("div", class_="_3pLy-c row")

    for laptop in laptop_divs:
        title = laptop.find("div", class_="_4rR01T").text
        price = laptop.find("div", class_="_30jeq3 _1_WHN1").text
        laptops_data.append({"title": title, "price": price})
```

```python
for i in range(20,31):
    page_start = time.time()
    scrape_page(i,BASE_URL)
    page_finish = time.time()
    print("Page {} took {} seconds to scrape".format(i,page_finish-page_start))
```

```python
df = pd.DataFrame(laptops_data)
df
```

```python
df.to_csv('laptop_price_flipkart.csv', index=False)
```

# 2) Covid data sraper


```python
COVID_URL = 'https://www.worldometers.info/coronavirus/'
```

```python
req = requests.get(COVID_URL)
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
thead = soup.find('thead').find_all('th')
COL_NAMES = [th.text.replace(u'\xa0',u' ').replace(u'\n','') for th in thead]
COL_NAMES = COL_NAMES[1:15]
COL_NAMES
```

```python
covid_data = []
```

```python
tbody = soup.find('tbody').find_all('tr')

for row in tbody:
    row_data = row.find_all('td')
    row_data = [elem.text for elem in row_data][1:15]
    row_data = dict(zip(COL_NAMES,row_data))
    covid_data.append(row_data)
```

```python
covid_data=covid_data[237:] #idk why but had to do this as it repeated North America 236 times :/
df = pd.DataFrame(covid_data)
df
```

```python
df.to_csv('covid_data.csv', index=False)
```
